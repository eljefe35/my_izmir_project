```python
import pandas as pd
import plotly.express as px
import os
from pathlib import Path
from dotenv import load_dotenv
```

```python
# Read the Mapbox API key
load_dotenv()
map_box_api = os.getenv("mapbox")
```

```python
# Set the Mapbox API
px.set_mapbox_access_token(map_box_api)
```

```python
# Read in data
acil_toplanma = pd.read_csv(Path('izmir_final__4.csv')).dropna()
```

```python
acil_toplanma.drop(columns=['Mahalle / Adres'])
```

```python
acil_toplanma['Enlem'] = acil_toplanma['Enlem'].astype(float)
acil_toplanma['Boylam'] = acil_toplanma['Boylam'].astype(float)
```

```python
latitude = acil_toplanma["Enlem"]
longitude = acil_toplanma["Boylam"]
```

```python
 # Plot Data
map = px.scatter_mapbox(
    acil_toplanma,
    title='Izmir Acil Toplanma Alanlari',
    lat=latitude,
    lon=longitude,
    hover_name="Mahalle / Adres",
    hover_data=["Kapasite", "Alan"],
    #text="Mahalle / Adres",
    color="Ilce",
    mapbox_style="outdoors",
    #size="Kapasite",
    zoom=8,
    width=1350,
    height=800
)

# Display the map
map.show()
```
